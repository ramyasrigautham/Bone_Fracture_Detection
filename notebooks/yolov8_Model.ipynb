```python
import glob
import os
file_sizes_dict = {}
for file in glob.glob(os.path.join(TRAIN_DATA_PATH["labels"], '*.txt')):
    #print(file)
    size = os.path.getsize(file)
    file_sizes_dict[size] = file_sizes_dict.setdefault(size, 0) + 1  # Corrected line

print(file_sizes_dict)
```

```python
def ExtractData(data_path: dict):
  data = {"File_Name" : [],
        "File_Size" : [],
        "Class" : [],
        "Cordinate": []
        }
  for file in glob.glob(os.path.join(data_path["labels"], '*.txt')):
    file_size = os.path.getsize(file)
    data["File_Name"].append(os.path.basename(file))
    data["File_Size"].append(file_size)
    if file_size == 0:
      data["Class"].append(-1)
      data["Cordinate"].append(None)
    else:
      with open(file, 'r') as f:
        line = f.readline()
        #print(line)
        data["Class"].append(line[0])
        data["Cordinate"].append(str(line[2:]))
        f.close()
  return data
```

```python
test_data_label_detail = ExtractData(TEST_DATA_PATH)
df_test = pd.DataFrame(test_data_label_detail)
df_test['Class'] = df_test['Class'].astype('int64')
```

```python
df_test = df_test[df_test['Class']!=-1]
```

```python
count=0
total=0
for i, r in enumerate(results):
    # Plot results image
    result = r
    file_name = result.path.split('/')[-1]
    cleaned_file_name = file_name.replace('jpg','txt')
    actual_classid= df_test.loc[df_test['File_Name']==cleaned_file_name]['Class']
    if(len(result.boxes)==0):
        if len(actual_classid)==0:
            count+=1
            print("Object: "+str(cleaned_file_name)+" : None  | Actual : None ")
            #print("Coordinates:", cords)
            print("Probability: 0")
        else:
            actual = result.names[actual_classid.item()]
            print("Object: "+str(cleaned_file_name)+" : None  | Actual :" + str(actual))
            print("Probability: 0")
    for box in result.boxes:
        class_id = result.names[box.cls[0].item()]
        if(len(actual_classid)>0):
            actual = result.names[actual_classid.item()]
        else:
            actual = 'None'

        cords = box.xyxy[0].tolist()
        cords = [round(x) for x in cords]
        conf = round(box.conf[0].item(), 2)
        print("Object: "+str(cleaned_file_name)+" : " +str(class_id)+" | Actual :" + str(actual))
        #print("Coordinates:", cords)
        print("Probability:", conf)
        if(actual ==class_id ):
            count=count+1
        #print("Actual result:", result.names[df_test.loc[df_test['File_Name']==cleaned_file_name]['Class'].item()])
        #print("---")
    total=total+1
print(count)
print(total)
```
